```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("preprocessed.csv")
```

```python
fig, ax = plt.subplots()
u=data[data["Died_year"]==2020]
u['Died_month'].hist(bins=12, linewidth=0.5, edgecolor="black")
ax.set_xlabel('Month')
ax.set_ylabel('Count')
ax.set_title('People died in 2020')
plt.show()
```

```python
fig, ax = plt.subplots()
u=data[data["Died_year"]==2021]
u['Died_month'].hist(bins=12, linewidth=0.5, edgecolor="black")
ax.set_xlabel('Month')
ax.set_ylabel('Count')
ax.set_title('People died in 2021')

# Tweak spacing to prevent clipping of ylabel
plt.show()
```

```python
fig, ax = plt.subplots()
u=data[(data["Died_year"]==2020) & (data["CLASIFFICATION"]=='Positive')]
u['Died_month'].hist(bins=12, linewidth=0.5, edgecolor="black")
ax.set_xlabel('Month')
ax.set_ylabel('Count')
ax.set_title('People died in 2020 with covid')
plt.show()
```

```python
fig, ax = plt.subplots()
u=data[(data["Died_year"]==2021) & (data["CLASIFFICATION"]=='Positive')]
u['Died_month'].hist(bins=12, linewidth=0.5, edgecolor="black")
ax.set_xlabel('Month')
ax.set_ylabel('Count')
ax.set_title('People died in 2021 with covid')
plt.show()
```

```python
u=data.loc[(data["Died_year"]!=1990)]
ax = sns.countplot(x ='Died_month',hue='CLASIFFICATION', data = u)
#plt.bar_label(ax.containers[0])
plt.title(" Died people count based on having covid or not   ",color="red");
```

```python
u=data.loc[(data["CLASIFFICATION"]=='Positive')]
ax = sns.countplot(x ='SEX', data = u)
plt.title(" + covid count for males vs females  ",color="red");
```

```python
u=data.loc[(data["CLASIFFICATION"]=='Positive')]
ax = sns.countplot(x ='DEATH', data = u)
plt.title(" Count of Died and alive + Covid  ",color="red");
```

```python
labels = ['Back-home','hospitalization']
u = [data['PATIENT_TYPE'].value_counts()[1],data['PATIENT_TYPE'].value_counts()[2]]
figure,axis = plt.subplots(figsize = (10,5))
axis.pie(u,labels = labels,explode = [0.05]*2,autopct='%20.0f%%',colors = sns.color_palette('twilight')[0:2])
plt.title('type of care',fontsize = 15)
plt.show()
```

```python
import plotly.graph_objects as go
b=np.arange(1,102)
layout = dict(title = 'Count of Died +covid having DISEASES for each Age',
              xaxis = dict(title = 'Age'),
              yaxis = dict(title = 'Count')
             )
fig = go.Figure(layout=layout)
fig.add_trace(go.Scatter(    x = b,
                              y = data[(data["DEATH"]=='Dead') & (data["CLASIFFICATION"]=='Positive')& (data["TOBACCO"]==1)].groupby('AGE').size(),
                             name = 'TOBACCO'))
fig.add_trace(go.Scatter(    x = b,
                              y = data[(data["DEATH"]=='Dead') & (data["CLASIFFICATION"]=='Positive')& (data["DIABETES"]==1)].groupby('AGE').size(),
                             name = 'DIABETES'))
fig.add_trace(go.Scatter(    x = b,
                              y = data[(data["DEATH"]=='Dead') & (data["CLASIFFICATION"]=='Positive')& (data["HIPERTENSION"]==1)].groupby('AGE').size(),
                             name = 'HIPERTENSION'))
fig.add_trace(go.Scatter(    x = b,
                              y = data[(data["DEATH"]=='Dead') & (data["CLASIFFICATION"]=='Positive')& (data["OBESITY"]==1)].groupby('AGE').size(),
                             name = 'OBESITY'))
fig.add_trace(go.Scatter(    x = b,
                              y = data[(data["DEATH"]=='Dead') & (data["CLASIFFICATION"]=='Positive')& (data["ASTHMA"]==1)].groupby('AGE').size(),
                             name = 'ASTHMA'))
fig.add_trace(go.Scatter(    x = b,
                              y = data[(data["DEATH"]=='Dead') & (data["CLASIFFICATION"]=='Positive')& (data["RENAL_CHRONIC"]==1)].groupby('AGE').size(),
                             name = 'RENAL_CHRONIC'))
fig.add_trace(go.Scatter(    x = b,
                              y = data[(data["DEATH"]=='Dead') & (data["CLASIFFICATION"]=='Positive')& (data["COPD"]==1)].groupby('AGE').size(),
                             name = 'COPD'))
fig.add_trace(go.Scatter(    x = b,
                              y = data[(data["DEATH"]=='Dead') & (data["CLASIFFICATION"]=='Positive')& (data["INMSUPR"]==1)].groupby('AGE').size(),
                             name = 'INMSUPR'))
fig.add_trace(go.Scatter(    x = b,
                              y = data[(data["DEATH"]=='Dead') & (data["CLASIFFICATION"]=='Positive')& (data["CARDIOVASCULAR"]==1)].groupby('AGE').size(),
                             name = 'CARDIOVASCULAR'))
fig.show()
```

```python
def countplot_diseases(Col_name):
    u=data.loc[(data[Col_name]==1)]
    ax = sns.countplot(x = "CLASIFFICATION", hue = "DEATH", data = u)
    plt.title(" People with "+Col_name,color="red");
```
